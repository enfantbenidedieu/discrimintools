## STEPDISC LDA - heart dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=8):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

### heart dataset

```python
#vins dataset
from discrimintools.datasets import load_heart
D = load_heart("train")
(
    GT(D.head().rename_axis("").reset_index())
    .tab_header(title=html("<b>Alcools dataset - training data</b>"))
)
```

### `instanciation & training`

```python
from discrimintools import DISCRIM
#split into X and y
y, X = D["disease"], D.drop(columns=["disease"])
clf = DISCRIM()
clf.fit(X,y)
```

### `stepdisc`

```python
from discrimintools import STEPDISC
clf2 = STEPDISC(method="backward",alpha=0.01,verbose=True)
clf2.fit(clf)
```

#### Selected variables

```python
#selected variables
print(clf2.summary_.selected)
```

### `summary`

```python
from discrimintools import summarySTEPDISC
```

#### Simple summary

```python
summarySTEPDISC(clf2)
```

#### Detailed summary

#### Markdown format
